```python
import datetime
import json

try:
    with open('pacientes.json', 'r') as f:
        pacientes = json.load(f)
except FileNotFoundError:
    pacientes = {}

try:
    with open('agendamentos.json', 'r') as f:
        agendamentos = json.load(f)
except FileNotFoundError:
    agendamentos = []

def cadastrar_paciente():
    nome = input("Digite o nome do paciente: ")
    telefone = input("Digite o telefone do paciente: ")
    if telefone in pacientes:
        print("Paciente já cadastrado!")
        return
    pacientes[telefone] = nome
    print("Paciente cadastrado com sucesso")

def marcar_consulta():
    print("Pacientes cadastrados:")
    for i, paciente in enumerate(pacientes.items(), start=1):
        print(f"{i}. {paciente[1]}")
    paciente_escolhido = int(input("Escolha o número do paciente: ")) - 1
    telefone_paciente = list(pacientes.keys())[paciente_escolhido]
    data = input("Digite a data da consulta (dd/mm/aaaa) utilize as barras: ")
    hora = input("Digite a hora da consulta (HH:MM) utilize os pontos: ")
    especialidade = input("Digite a especialidade desejada: ")
    data_hora = datetime.datetime.strptime(data + ' ' + hora, '%d/%m/%Y %H:%M')
    if data_hora < datetime.datetime.now():
        print("Não é possível agendar consultas retroativas.")
        return
    for agendamento in agendamentos:
        if agendamento['data_hora'] == data_hora.isoformat():
            print("Horário já agendado.")
            return
    agendamentos.append({'telefone': telefone_paciente, 'data_hora': data_hora.isoformat(), 'especialidade': especialidade})
    print("Consulta agendada com sucesso")

def cancelar_consulta():
    print("Agendamentos:")
    for i, agendamento in enumerate(agendamentos, start=1):
        print(f"{i}. {pacientes[agendamento['telefone']]} - {datetime.datetime.fromisoformat(agendamento['data_hora']).strftime('%d/%m/%Y %H:%M')} - {agendamento['especialidade']}")
    agendamento_escolhido = int(input("Escolha o número do agendamento que deseja cancelar: ")) - 1
    agendamentos.pop(agendamento_escolhido)
    print("Agendamento cancelado com sucesso")

def menu():
    while True:
        print("\n1. Cadastrar paciente")
        print("2. Marcar consulta")
        print("3. Cancelar consulta")
        print("4. Sair")
        opcao = int(input("Escolha uma opção: "))
        if opcao == 1:
            cadastrar_paciente()
        elif opcao == 2:
            marcar_consulta()
        elif opcao == 3:
            cancelar_consulta()
        elif opcao == 4:
            break
    with open('pacientes.json', 'w') as f:
        json.dump(pacientes, f)
    with open('agendamentos.json', 'w') as f:
        json.dump(agendamentos, f)

menu()
```
